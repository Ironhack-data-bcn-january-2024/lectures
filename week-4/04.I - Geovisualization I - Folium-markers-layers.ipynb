# Festivos vs Holiday vs vispera

````python
"""
for index, row in df.iterrows():
    print(f"Index: {index}")
    print(f"Row: {row}")
"""
````

````python
df["festividad"].unique()
````

````python
mad_lat = 40.4166852
mad_lon = -3.705739617

df.sample()

# COLUMNS: lon, lat, festividad

mad_map = Map(location =[mad_lat, mad_lon], zoom_start = 15)

for index, row in df.iterrows():
    
    festividad = row["festividad"]
    lat_accident = row["lat"]
    lon_accident = row["lon"]
    
    if festividad == "Vispera":
        color = "yellow"
        icon_color = "black"
        icon = "beer"
    
    elif festividad == "Laborable":
        color = "white"
        icon_color = "red"
        icon = "frown-o"

    elif festividad == "Festivo":
        color = "green"
        icon_color = "white"
        icon = "plane"

    icon = Icon(
                color = color,
                icon_color = icon_color,
                icon = icon,
                prefix="fa" # font-awesome website: fa
        )
    
    marker = Marker(location = [lat_accident, lon_accident], icon = icon, tooltip=festividad)
    marker.add_to(mad_map)
    
mad_map
````

````python
mad_lat = 40.4166852
mad_lon = -3.705739617


mad_map = Map(location =[mad_lat, mad_lon], zoom_start = 15)

for index, row in df.iterrows():

    all_together = {"festividad": row["festividad"],
                    "location": [row["lat"],row["lon"]],
                    "tooltip": row["festividad"]}

    if all_together["festividad"] == "Vispera":
        color = "yellow"
        icon_color = "black"
        icon = "beer"

    elif all_together["festividad"] == "Laborable":
        color = "white"
        icon_color = "red"
        icon = "frown-o"

    elif all_together["festividad"] == "Festivo":
        color = "green"
        icon_color = "white"
        icon = "plane"

    icon = Icon(
                color = color,
                icon_color = icon_color,
                icon = icon,
                prefix="fa" # font-awesome website: fa
    )

marker = Marker(**all_together, icon = icon) # This will look up the keys in the dictionary
# and pass the values as the arguemnts the Marker receives

marker.add_to(mad_map) 

mad_map
````
