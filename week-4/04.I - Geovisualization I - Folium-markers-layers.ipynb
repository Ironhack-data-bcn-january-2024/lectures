````python
mad_lat = 40.4166852
mad_lon = -3.705739617

df.sample()

#Â COLUMNS: lon, lat, festividad

mad_map = Map(location =[mad_lat, mad_lon], zoom_start = 15)

for index, row in df.iterrows():
    
    festividad = row["festividad"]
    lat_accident = row["lat"]
    lon_accident = row["lon"]
    
    if festividad == "Vispera":
        color = "yellow"
        icon_color = "black"
        icon = "beer"
    
    elif festividad == "Laborable":
        color = "white"
        icon_color = "red"
        icon = "frown-o"

    elif festividad == "Festivo":
        color = "green"
        icon_color = "white"
        icon = "plane"

    icon = Icon(
                color = color,
                icon_color = icon_color,
                icon = icon,
                prefix="fa" # font-awesome website: fa
        )
    
    marker = Marker(location = [lat_accident, lon_accident], icon = icon, tooltip=festividad)
    marker.add_to(mad_map)
    
mad_map
````

````python
mad_lat = 40.4166852
mad_lon = -3.705739617


mad_map = Map(location =[mad_lat, mad_lon], zoom_start = 15)

for index, row in df.iterrows():

    all_together = {"festividad": row["festividad"],
                    "location": [row["lat"],row["lon"]],
                    "tooltip": row["festividad"]}

    if all_together["festividad"] == "Vispera":
        color = "yellow"
        icon_color = "black"
        icon = "beer"

    elif all_together["festividad"] == "Laborable":
        color = "white"
        icon_color = "red"
        icon = "frown-o"

    elif all_together["festividad"] == "Festivo":
        color = "green"
        icon_color = "white"
        icon = "plane"

    icon = Icon(
                color = color,
                icon_color = icon_color,
                icon = icon,
                prefix="fa" # font-awesome website: fa
    )

marker = Marker(**all_together, icon = icon) #Â This will look up the keys in the dictionary
#Â and pass the values as the arguemnts the Marker receives

marker.add_to(mad_map) 

mad_map
````

##  Morning bikers will be safer  ðŸš´

Let's check how many categories there are in the time column, to make a heat map. First we make a map with the morning schedule, and then we add the afternoon schedule with the Layer Control.
Here we go ðŸ”¥

````python
df.sample()
````

````python
df["horario"].value_counts()
````

### We create the basic map to which we then add the layers

````python
mad_map = Map(location=[mad_lat, mad_lon], zoom_start=15)
````

### Subset the dataframe per layer

````python
morning = df[df["horario"] == "MaÃ±ana"] #Â subset
evening = df[df["horario"] == "Tarde"]
night = df[df["horario"] == "Noche"]
````

### Create the HeatMap layer and add it to the FeatureGroup

````python
#Â 1. Susbet
morning = df[df["horario"] == "MaÃ±ana"]

#Â 2. Feature group
morning_group = folium.FeatureGroup(name=f"Morning: {morning.shape[0]}")

#Â 3. Create a HeatMap based on the feature group
HeatMap(data = morning[["lat", "lon"]]).add_to(morning_group) #only the lat,lon from the mornings

# 4. Add the group to the map
morning_group.add_to(mad_map)
````

````python
evening = df[df["horario"] == "Tarde"]
evening_group = folium.FeatureGroup(name=f"Evening: {evening.shape[0]}")
HeatMap(data = evening[["lat", "lon"]]).add_to(evening_group) #only the lat,lon from the evenings
evening_group.add_to(mad_map)
````

````python
night = df[df["horario"] == "Noche"]
night_group = folium.FeatureGroup(name=f"Night: {night.shape[0]}")
HeatMap(data = night[["lat", "lon"]]).add_to(night_group) #only the lat,lon from the nights
night_group.add_to(mad_map)
````

````python
folium.LayerControl(collapsed=False, position="topleft").add_to(mad_map);
````
