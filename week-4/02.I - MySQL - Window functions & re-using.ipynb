#Â Window Functions

These functions provide powerful tools for analyzing and processing data within specific partitions or frames, enabling advanced analytics and reporting capabilities in MySQL.

To use window functions in MySQL, you typically include the function along with the OVER() clause, which defines the partitioning, ordering, and window frame for the function. This allows you to calculate aggregates, rankings, and other analytical insights over specific groups of rows in the result set.

#Â Aggregate Functions:

- `SUM()`: Calculates the sum of a specified column over a defined window.
- `AVG()`: Computes the average of a specified column within a given window.
- `COUNT()`: Counts the number of rows in a specified window.
- `MIN()`: Finds the minimum value in a specified column over a window.
- `MAX()`: Finds the maximum value in a specified column over a window.

```sql
SELECT
    Product,
    Amount,
    SUM(Amount) OVER () AS TotalSales,
    AVG(Amount) OVER () AS AvgSales,
    COUNT(*) OVER () AS TotalRows,
    MIN(Amount) OVER () AS MinSale,
    MAX(Amount) OVER () AS MaxSale
FROM
    Sales;
```

# Ranking Functions

- `DENSE_RANK()`: Assigns a dense rank (no gaps) to each row in a window.

SELECT

#Â Analytic Functions

- `LEAD():` Accesses data from subsequent rows within a window.
- `LAG():` Accesses data from preceding rows within a window.
- `FIRST_VALUE():` Returns the first value in a window based on the specified order.
- `LAST_VALUE():` Returns the last value in a window based on the specified order.

```sql
SELECT
    Product,
    Amount,
    LEAD(Amount) OVER (ORDER BY Amount) AS NextSaleAmount,
    LAG(Amount) OVER (ORDER BY Amount) AS PrevSaleAmount,
    FIRST_VALUE(Amount) OVER (PARTITIOM BY Amount) AS FirstSaleAmount,
    LAST_VALUE(Amount) OVER (PARTITION BY Amount) AS LastSaleAmount
FROM
    Sales;
```

## Temporary Tables

Temporary tables are staging tables that are built from the results of queries and dropped after the database session has ended. They serve to store the results of the query for later use in the session. The way to avoid having to write the same subquery multiple times is to save the query results to a temporary table and then query the temporary table each time we need to access those results.

Let's see how we could have changed the subquery from the previous section to a temporary table. To do this, we can use the CREATE TEMPORARY TABLE command, followed by the name we want to give the new table and the query.
