````python
def find_things (venue, lat, lon):

    url = f"https://api.foursquare.com/v3/places/search?query={venue}&ll={lat}%2C{lon}&sort=DISTANCE"

    headers = {
        "accept": "application/json",
        "Authorization": token
    }
    
    params = {"limit":20}

    response = requests.get(url, headers=headers, params= params).json()

    return response

result_madrid = find_things ("bookstores", "40.42879","-3.68060")
result_madrid
````

````python
name = result_madrid["results"][10]["name"]
location = result_madrid["results"][10]["location"]["formatted_address"]
lat = result_madrid["results"][10]["geocodes"]["main"]["latitude"]
lon = result_madrid["results"][10]["geocodes"]["main"]["longitude"]
````

````python
new_list = []

for bookstore in result_madrid["results"]:

    name = bookstore["name"]
    location = bookstore["location"]["formatted_address"]
    lat = bookstore["geocodes"]["main"]["latitude"]
    lon = bookstore["geocodes"]["main"]["longitude"]
        
    dict_ = {
        "name": name,
        "location": location,
        "lat": lat,
        "lon": lon
    }
    
    new_list.append(dict_)
    
df = pd.DataFrame(new_list)
df
````

````python
def process_response_into_df (response):
    
    new_list = []

    for bookstore in response["results"]:

        name = bookstore["name"]
        location = bookstore["location"]["formatted_address"]
        lat = bookstore["geocodes"]["main"]["latitude"]
        lon = bookstore["geocodes"]["main"]["longitude"]

        dict_ = {
            "name": name,
            "location": location,
            "lat": lat,
            "lon": lon
        }

        new_list.append(dict_)

    return pd.DataFrame(new_list)
````
